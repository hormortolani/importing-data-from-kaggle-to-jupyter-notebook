```python
extracted_files = os.listdir('vehicle-population-statistics')
print('Extracted files:', extracted_files)
```

```python
def extract_start_year(year_range):
    start_year = year_range.split('-')[0]
    return start_year

df['StartYear'] = df['Year'].apply(extract_start_year)

df['StartYear'] = pd.to_datetime(df['StartYear'], format='%Y')

df['Year'] = df['StartYear']
df.drop(columns='StartYear', inplace=True)
```
